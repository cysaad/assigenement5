### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
data_csv_2020=pd.read_csv('resources/athletic_sales_2020.csv')
data_csv_2021=pd.read_csv('resources/athletic_sales_2021.csv')
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_data=pd.concat([data_csv_2020,data_csv_2021]).reset_index(drop=True)
combined_data
```

```python
# Check if any values are null.
null_values=combined_data.isnull().sum()
null_values
```

```python
# Check the data type of each column
data_types=combined_data.dtypes
data_types
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_data['invoice_date']=pd.to_datetime(combined_data['invoice_date'],format='%Y-%m-%d')
combined_data
```

```python
# Confirm that the "invoice_date" data type has been changed.
comfirmed_data_type=combined_data['invoice_date'].dtype
comfirmed_data_type
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_data=combined_data.groupby(['region','state','city'])['units_sold'].sum().reset_index(name='Total_Products_Sold')
top_five_results=grouped_data.sort_values('Total_Products_Sold',ascending=False).head(5)


# Show the top 5 results.
top_five_results
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = combined_data.pivot_table(
    index=['region', 'state', 'city'], 
    values='units_sold', 
    aggfunc='sum'
).rename(columns={'units_sold': 'Total_Products_Sold'})


# Rename the "units_sold" column to "Total_Products_Sold"
top_five = pivot_table.sort_values('Total_Products_Sold', ascending=False).head(5)


# Show the top 5 results.
top_five
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

pivot_table = combined_data.pivot_table(index='invoice_date', values='total_sales', aggfunc='sum')
# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
pivot_table
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = pivot_table.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=True)
sorted_daily_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = pivot_table.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sorted_weekly_sales = weekly_sales.sort_values(by='Total Sales', ascending=False)
sorted_weekly_sales 
```
